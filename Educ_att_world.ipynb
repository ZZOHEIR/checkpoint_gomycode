```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_educ_na = df_educ.dropna()
df_educ_na.isna().sum()
```

```python
df_educ_na = df_educ_na.drop_duplicates()
df_educ_na.info()
```

```python

df_groupby = df_educ_na.groupby(['country']).mean().reset_index()
df_groupby1 = df_groupby.drop(columns = 'year')
df_groupby1
```

```python
numeric_df = df_educ_na.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
```

```python


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()
```

```python
# filtre : essai avec annee 2015
df_educ_na_2015 = df_educ_na[df_educ_na['year'] == 2015]
df_educ_na_2015.reset_index(inplace = True, drop = True)
df_educ_na_2015.head()
```

```python
# visualiser les outliers avec sns.boxplot
for column in df_educ_na.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df_educ_na[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
plt.show()
```

```python
# travailler sur les Outliers :
for column in df_educ_na.columns:
    if df_educ_na[column].dtype in ['float64', 'int64']:  # Apply only to numeric columns
        Q1 = df_educ_na[column].quantile(0.25)
        Q3 = df_educ_na[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out outliers
        df_educ_na_new = df_educ_na[(df_educ_na[column] >= lower_bound) & (df_educ_na[column] <= upper_bound)]

        plt.figure(figsize=(10, 5))
        sns.boxplot(x=df_educ_na_new[column])
        plt.title(f'Boxplot of {column} after outlier removal')
        plt.show()
```
