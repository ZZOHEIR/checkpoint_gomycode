```python
#Loan_prediction_dataset google drive link : /content/drive/MyDrive/Colab_Notebooks/Loan_prediction_dataset.csv'
file = open('/content/drive/MyDrive/Colab_Notebooks/Loan_prediction_dataset.csv', 'r')
file

```

```python
from numpy import genfromtxt
data = genfromtxt('/content/drive/MyDrive/Colab_Notebooks/Loan_prediction_dataset.csv', dtype=None, delimiter=',', skip_header=1)
data
```

```python
loan_amounts = data['f8']


# Calculate mean, median, and standard deviation
mean_loan = np.mean(loan_amounts)
median_loan = np.median(loan_amounts)
std_loan = np.std(loan_amounts)

# Print the results
print(f"Mean Loan Amount: {mean_loan}")
print(f"Median Loan Amount: {median_loan}")
print(f"Standard Deviation of Loan Amounts: {std_loan}")
```
