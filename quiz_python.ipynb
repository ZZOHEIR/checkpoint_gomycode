```python
import random
import tkinter as tk
from tkinter import messagebox

# Define quiz questions
questions = [
    ("Quel composant de l’atome porte une charge électrique négative ?", "L’électron", ["Le proton", "L’électron", "Le neutron", "Le quark"]),
    ("Quelle est la spécialité d’un néphrologue ?", "Les reins", ["Le cœur", "Les poumons", "Les reins", "Le cerveau"]),
    ("Lequel de ces pays d’Asie est aujourd’hui gouverné par un roi ?", "Thaïlande", ["Japon", "Chine", "Thaïlande", "Vietnam"]),
    ("Quelle est la vitesse du son dans l’air ?", "343 m/s", ["343 m/s", "299,792 km/s", "1235 km/h", "7.9 km/s"]),
    ("Quelle surface de la Terre les eaux recouvrent-elles ?", "71%", ["50%", "71%", "80%", "90%"])
]

random.shuffle(questions)

score = 0

root = tk.Tk()
root.title("Interactive Python Quiz Game")

def display_question(index):
    if index < len(questions):
        question, correct_answer, options = questions[index]
        label_question.config(text=question)

        for i, option in enumerate(options):
            radio_buttons[i].config(text=option, value=option)

        validate_button.config(command=lambda: validate_answer(index, correct_answer))
    else:
        show_final_score()

# Fenetre de validation des entrees
def validate_answer(index, correct_answer):
    selected_answer = selected_option.get()
    if selected_answer == correct_answer:
        messagebox.showinfo("Validation", "Correct!")
        global score
        score += 1
    else:
        messagebox.showinfo("Validation", "Incorrect. La bonne réponse était: " + correct_answer)

    # aller vers le question suivante
    display_question(index + 1)

# afficher le resultat
def show_final_score():
    messagebox.showinfo("Score Final", f"Votre score est: {score}/{len(questions)}")
    root.quit()

# tkimter :
label_question = tk.Label(root, text="", justify='left')
label_question.pack(side='top', anchor='w')

#
selected_option = tk.StringVar()
radio_buttons = [tk.Radiobutton(root, text="", variable=selected_option, value="") for _ in range(4)]
for radio_button in radio_buttons:
    radio_button.pack(side='top', anchor='w')

validate_button = tk.Button(root, text='Valider')
validate_button.pack(side='top')

# Display the first question
display_question(0)

# Start the Tkinter event loop
root.mainloop()
```
